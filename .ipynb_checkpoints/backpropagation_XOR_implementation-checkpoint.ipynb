# Backpropagation Algorithm Explained

## Implemented using PyTroch

```python
#import functions
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
#defining the neural network of model traning
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        
        #defining the layer 
        self.fc1 = nn.Linear(2,2)
        self.fc2 = nn.Linear(2,1)
        
    def forward(self,x):
        x = torch.sigmoid(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x
    
#define the loss function and the optimizer
net = Net()
criterion = nn.MSELoss() 
optimizer = optim.SGD(net.parameters(),lr=0.1)

#Input and the output for the XOR function
x = torch.tensor([[1,1],[1,0],[0,1],[0,0]],dtype = torch.float32)
y = torch.tensor([[0],[1],[1],[0]],dtype = torch.float32)

#Train the network
for epoch in range(100000):
    optimizer.zero_grad()
    output = net(x)
    loss = criterion(output,y)
    loss.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print(f'Epoch{epoch},loss:{loss.item()}')
        
        
```

## Command to predict the algorithm output

```python

#input to the network 
test_input = torch.tensor([0,1],dtype = torch.float32)


with torch.no_grad():
    test_output = net(test_input)
    print(f'Predicted output:{test_output.item()}')  
    #simple implementation of sigmoid activation function
    if test_output.item()>0.7:  # if the confidence is greater than 0.7 
        print("the result is",1)
    else:
        print("the result is",0)
```
